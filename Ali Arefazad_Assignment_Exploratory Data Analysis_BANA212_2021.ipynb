** Create a new dataframe called returns. This dataframe will contain the returns for each company's stock. returns are typically defined by:**

$$r_t = \frac{p_t - p_{t-1}}{p_{t-1}} = \frac{p_t}{p_{t-1}} - 1$$

** We can use pandas pct_change() method on the Close column to create a new dataframe representing this return value. Use .groupby().**

```python
df2= pd.concat([AMZN,FB,MSFT,TWTR,AAPL,GOOG],axis=1)
df2
```

```python
returns=df2.Close.pct_change().set_axis(['AMZN', 'FB','MSFT','TWTR','AAPL','GOOG'],axis=1, inplace=False)
returns
```

** Using this returns DataFrame, figure out on what dates each company stock had the best and worst single day returns. Did anything significant happen that day?**

```python
returns.idxmax()
```

```python
# It looks like 4 companies(Facebook, Microsoft, Apple, and Google had best return in a same day 03/13/2020) which all universities 
# were informed that classes would be held virtually dye to pandemic and Covid-19 crisis.
```

```python
returns.idxmin()
```

```python
#Ironically, the same 4 company had their worst return 3 days after thier best return on 03/16/2020.
```

** Please state here what you have noticed. Did anything significant happen in that time frame? **

```python
# Stocks on Thursday 03/13/2020 suffered their deepest one-day percentage loss since the 1987 "Black Monday" 
#stock market crash, continuing to fall due to mounting fears of widespread economic damage from the coronavirus 
#pandemic. in those days everything was influenced by pandemic and Covid-19 crisis and all those flactuations
# were the result of Covid-19.
```

** Take a look at the standard deviation of the returns, which stock would you classify as the riskiest over the entire time period? **

```python
returns_std = returns.std()
returns_std
```

```python
# by considering the standard deviation, Twitter is the riskiest stock and Google alongside 
# Amazon have the least risk.
```

Which company stock would you classify as the most profitable for the year 2020?"

```python
#Which company stock would you classify as the most profitable for the year 2020?"
#Compute cumulative daily returns

cumulative_return = (1 + returns).cumprod() - 1
cumulative_return = cumulative_return.reset_index()
cumulative_return
```

```python
#Last record of the dataframe multiplied by 100 is giving us the percentage change of the stock prices for our entire period
cum_return_entire_period = cumulative_return.iloc[:, 1:].tail(1)
cum_return_entire_period * 100
```

```python
#These are the perlrcentage change in prices
# it is clear that Apple was the most profitable stock of 2020 by 78% return and Amazon had the second place.
# Also, it is shown that Google was had the least return alongside facebook.
```
